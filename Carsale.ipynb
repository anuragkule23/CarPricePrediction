```python
X=car[['Car_Name','Year','Kms_Driven','Fuel_Type']]
y=car['Selling_Price']
```

```python
X
```

```python
y.shape
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.2)
```

```python
from sklearn.linear_model import LinearRegression
```

```python
from sklearn.preprocessing import OneHotEncoder
from sklearn.compose import make_column_transformer
from sklearn.pipeline import make_pipeline
from sklearn.metrics import r2_score
```

```python
ohe=OneHotEncoder()
ohe.fit(X[['Car_Name','Fuel_Type']])
```

```python
column_trans=make_column_transformer((OneHotEncoder(categories=ohe.categories_),['Car_Name','Fuel_Type']),
                                    remainder='passthrough')
```

```python
lr=LinearRegression()
```

```python
pipe=make_pipeline(column_trans,lr)
```

```python
pipe.fit(X_train,y_train)
```

```python
y_pred=pipe.predict(X_test)
```

```python
r2_score(y_test,y_pred)
```

```python
scores=[]
for i in range(1000):
    X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.1,random_state=i)
    lr=LinearRegression()
    pipe=make_pipeline(column_trans,lr)
    pipe.fit(X_train,y_train)
    y_pred=pipe.predict(X_test)
    scores.append(r2_score(y_test,y_pred))
```

```python
np.argmax(scores)
```

```python
scores[np.argmax(scores)]
```

```python
pipe.predict(pd.DataFrame(columns=X_test.columns,data=np.array(['ritz',2014,27000,'Petrol']).reshape(1,4)))
```

```python
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.1,random_state=np.argmax(scores))
lr=LinearRegression()
pipe=make_pipeline(column_trans,lr)
pipe.fit(X_train,y_train)
y_pred=pipe.predict(X_test)
r2_score(y_test,y_pred)
```
